### 엔트로피

- 열역학 용어로 무질서의 정도를 의미
- p는 해당 데이터가 해당 클래스에 속할 확률, 값이 커질수록 불확실성 커짐

$$
-p_i \log_2(p_i)
$$

$$
Entropy = \sum_{k=1}^{m} -p_i \log_2(p_i)
$$


```python
import numpy as np

p = np.arange(0.001, 1 , 0.001)
plt.plot(p, -p*np.log2(p))
```

```python
(-(10/16)*np.log2(10/16)) - ((6/16) * np.log2(6/16))
```

```python
0.5 * ((-(7/8)*np.log2(7/8)) - (1/8) * np.log2(1/8)) + 0.5 * ((-(3/8)*np.log2(3/8)) - (5/8) * np.log2(5/8))
```

### 지니계수

- Gini index 혹은 불순도율
- 엔트로피의 계산량이 맣아서 비슷한 개념이면서 보다 계산량이 적은 지니계수를 사용하는 경우가 많다


$$
Gini = \sum_{k=1}^{d} R_i \left(1 - \sum_{k=1}^{m} p_{ik}^2 \right)
$$

```python
1-(6/16)**2 - (10/16)**2
```

```python
0.5 * ( 1 - (7/8)**2 - (1/8)**2) + 0.5*( 1 - (3/8)**2 - (5/8)**2)
```

### sklearn을 이용한 결정나무의 구현

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
```

### accuracy 확인

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```

### plot_decision_regions : 학습을 완료한 모델을 준다 (clf)

- 과적합(overfitting) : 일반화할 수 x , 내가 준 데이터에 너무 최적화 되서 불확실성을 가짐
    - 기계학습에서 학습 데이터를 과하게 학습
    - 모델이 over-fitting되면, train-set에서는 정확도가 매우 높게 나오지만 test-set에서는 낮은 정확도가 나온다. 한마디로 범용성이 없는 모델이 된다.

- 과적합 해결방법
    - 데이터를 추가 수집한다.
    - feature의 개수를 줄인다.
    - Regularization
    - Early Stopping
    - Model Averaging
    - Dropout

- 추론기 : 학습을 완료한 모델을 사용

- 데이터 나누기 : train_test_split()
- stratify : label의 비율을 맞춰준다
- fit -> 훈련용 데이터에 대해 train set
- test data set은 훈려에 참가 x , accuracy 측정

```python
from sklearn.datasets import load_iris
import pandas as pd 

iris = load_iris()
```

### 데이터를 훈련 / 테스트로 분리

```python
from sklearn.model_selection import train_test_split

features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size=0.2,
                                                    random_state=13)
```

### 훈련용 / 테스트용 데이터 분리 확인

- 옵션 : stratify

```python
import numpy as np 

np.unique(y_test, return_counts=True)
```

```python
features = iris.data
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size=0.2,
                                                    stratify=labels,
                                                    random_state=13)
```

```python
np.unique(y_test, return_counts=True)
```

### train 데이터만 대상으로 결정나무 모델 만들기

```python
iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

### train 데이터에 대한 accuracy 확인

```python
y_pred_tr  = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

### 모델 확인

```python
import matplotlib.pyplot as plt 
from sklearn import tree 

fig = plt.figure(figsize = (15,8))
_= tree.plot_tree(iris_tree,
                   feature_names= list(iris.feature_names),
                   class_names = list(iris.target_names),
                   filled = True)
```

### 훈련데이터에 대한 결정경계 확인

```python
plt.figure(figsize = (12,8))
plot_decision_regions(X = X_train, y=y_train, clf = iris_tree, legend=2)
plt.show()
```

### 테스트 데이터에 대한 accuracy

```python
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

### 전체 데이터에서 관찰해보기

```python
scatter_highlight_kwargs = {'s' : 150, 'label' : 'Test data', 'alpha' : 0.9}
scatter_kwargs = {'s' : 120, 'edgecolor' : None , 'alpha' : 0.7}

plt.figure(figsize = (12,8))
plot_decision_regions(X = features, y = labels,
                      X_highlight=X_test, clf = iris_tree, legend=2,
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs=scatter_kwargs,
                      contourf_kwargs={'alpha' : 0.2})
plt.show()
```
